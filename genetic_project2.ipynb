```python
#lets find and clear null values
def null_clear(data,columns):
    for col in columns:
        null_counter=data[col].isnull().sum()
        if null_counter>0:
             data[col]=data[col].fillna(data[col].mean())
        else:
             print(f"values not null in {col}")

column_list=list(data.columns)
null_clear(data,column_list)
```

```python
#handle outliers
def check_and_handle_outliers(column,df):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)

  IQR= Q3 - Q1

  lower_limit = Q1 - 1.5 * IQR
  upper_limit = Q3 + 1.5 * IQR

  for rows in range(len(df)):
    if df.loc[rows,column] > upper_limit:
      df.loc[rows,column] = upper_limit

    elif df.loc[rows,column] < lower_limit:
      df.loc[rows,column] = lower_limit
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from collections import Counter
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,random_state=40)
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
logr=LogisticRegression()
logr.fit(X_train,Y_train)
```

```python
y_pred=logr.predict(X_test)
print("train score : ",logr.score(X_train,Y_train)*100)
print("Test score : ",logr.score(X_test,Y_test)*100)
```

```python
# Function to calculate accuracy
def evaluate_features(individual):
    selected_features = [bool(x) for x in individual]
    X_train_selected = X_train[:, selected_features]
    X_test_selected = X_test[:, selected_features]

    logr.fit(X_train_selected, Y_train)
    y_pred = logr.predict(X_test_selected)

    return accuracy_score(Y_test, y_pred),
```

```python
# Define custom crossover operation
def custom_crossover(ind1, ind2):
    size = min(len(ind1), len(ind2))
    cxpoint = np.random.randint(1, size)
    ind1[cxpoint:], ind2[cxpoint:] = ind2[cxpoint:], ind1[cxpoint:]
    return ind1, ind2
```

```python
# Define custom mutation operation
def custom_mutation(individual, indpb):
    for i in range(len(individual)):
        if np.random.random() < indpb:
            individual[i] = type(individual[i])(not individual[i])
    return individual,
```

```python
from deap import base, creator, tools, algorithms
# Define the genetic algorithm parameters
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_bool", np.random.choice, [0, 1])
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=len(X.columns))
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("evaluate", evaluate_features)
toolbox.register("mate",  tools.cxBlend, alpha=0.5)
toolbox.register("mutate",tools.mutGaussian, mu=0, sigma=0.2, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)
```

```python
# Genetic Algorithm
population = toolbox.population(n=50)
algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.2, ngen=10, verbose=False)
```

```python
# Retrieve the best individual and selected features
best_individual = tools.selBest(population, k=1)[0]
selected_features = [bool(x) for x in best_individual]
```

```python
# Train and test the model with the selected features
X_train_selected = X_train[:, selected_features]
X_test_selected = X_test[:, selected_features]
```

```python
logr.fit(X_train_selected, Y_train)
y_pred = logr.predict(X_test_selected)

print("Selected Features:", X.columns[selected_features])
print("Train Score:", logr.score(X_train_selected, Y_train) * 100)
print("Test Score:", logr.score(X_test_selected, Y_test) * 100)
```
